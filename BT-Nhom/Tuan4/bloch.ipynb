```python
def RK4_he(t_i, y_i, dt, phuongtrinh):

    k1 = dt * phuongtrinh(t_i, y_i)
    k2 = dt * phuongtrinh(t_i + dt/2, y_i + k1/2)
    k3 = dt * phuongtrinh(t_i + dt/2, y_i + k2/2)
    k4 = dt * phuongtrinh(t_i + dt,   y_i + k3)

    y_moi = y_i + (k1 + 2*k2 + 2*k3 + k4)/6
    return y_moi
```
