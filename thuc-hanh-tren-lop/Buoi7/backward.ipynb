```python
def hamso(kappa,C, rho, N_x, N_time,l,t_max):
    h = l/(N_x-1)
    k = t_max/(N_time-1)
    alpha = np.sqrt(kappa/(C*rho))
    eta  = alpha ** 2 * k / h**2

    if eta <= 0.5:
        return eta,h,k
    else:
        return False
```
