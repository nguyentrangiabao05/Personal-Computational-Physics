## Bai 1: Giai phuong trinh f(x)

### Phuong phap bisection

1) Cho a va b, f(x). Va f(a) be hon f(b)
2) c = (a+b)/2
3) f(c) > 0 thi cho b = c
4) Tinh do chinh xac

5) Co 2 cach chon dieu kien la delta f < 0, c_n+1 - c_n < 0





```python
a = -5
b = 5
epsilon = 1e-5
N = 100

def hamso(x):
    y = x + 2
    return y
i = 1

def bisection(a,b,epsilon,N):
    for i in range(0,N):
        c = (a+b)/2
        if hamso(b)*hamso(c) > 0:
            b = c
        else:
            a = c
        if abs(hamso(c))<epsilon:
            print('\n Root found with eps = ', c)
            break
        if i == N -1:
            print('\n Root NOT found after Nmax interation')    
        i = i + 1

bisection(a,b,epsilon,N)
```

```python
import numpy as np

def hamso1(x):
    y = x - 2**(-x)
    return y
def hamso2(x):
    y = np.exp(x) - 2 -np.cos(np.exp(x)-2)
    return y

def bisection_file(a,b,epsilon,N,filename,hamso):
    with open(filename, "w", encoding="utf-8") as file:
        file.write("Giai phuong trinh f(x) = 0 bang bisection\n")
        file.write("#" * 70 + "\n")
        file.write(f"{'i':>3s}  {'a':>12s}  {'b':>12s}  {'c':>16s}  {'f(c)':>18s}\n")
        for i in range(0,N):
            c = (a+b)/2
            fb = hamso(b)
            fc = hamso(c)
            if fb*fc > 0:
                b = c
            else:
                a = c
            file.write(f"{i:3d}  {a:12.6f}  {b:12.6f}  {c:16.10f}  {fc:18.10e}\n")
            if abs(fc)<epsilon:
                print('\n Root found with eps = ', c)
                break
            if i == N -1:
                print('\n Root NOT found after Nmax interation')    
            i = i + 1

def bisection_print(a,b,epsilon,N,hamso):
        print("Giai phuong trinh f(x) = 0 bang bisection\n")
        print("#" * 70 + "\n")
        print(f"{'i':>3s}  {'a':>12s}  {'b':>12s}  {'c':>16s}  {'f(c)':>18s}\n")
        for i in range(0,N):
            c = (a+b)/2
            fb = hamso(b)
            fc = hamso(c)
            if fb*fc > 0:
                b = c
            else:
                a = c
            print(f"{i:3d}  {a:12.6f}  {b:12.6f}  {c:16.10f}  {fc:18.10e}\n")
            if abs(fc)<epsilon:
                print('\n Root found with eps = ', c)
                break
            if i == N -1:
                print('\n Root NOT found after Nmax interation')    
            i = i + 1

filename = 'ketquatinh_bisection'
a = -5
b = 5
epsilon = 1e-10
N = 100

bisection_print(a,b,epsilon,N,hamso2)
bisection_file(a,b,epsilon,N,filename,hamso2)
```

## Bai 2: Phuong phap diem co dinh

### Dinh nghia

p la diem co dinh cua ham g(x) da cho neu p =g(x)

Gia su nen p la nghiem cu f(x), f(p) = 0, ta co the chon ham g(x) voi 1 diem co dinh p theo nhieu cach

VD g(x) = x - f(x) hoac g(x) = x - 6 f(x)

The p vao cac ham sau thi ta se thu duoc: g(p) = p

Khi do neu g(x) co 1 diem co dinh o p thi ham f duoc cho boi f(x)=x-g(x) co diem co diem zero tai p, tuc p la nghiem cua f(x)

### Dinh ly

Neu g lien tuc tren [a,b] g thuoc [a,b] 

### Lap de tim diem co dinh

1) Chon truoc p_0 ban dau, tao chuoi pn = g(pn-1), n>=1

2) Neu chuoi hoi tu ve p va (g lien tuc) thi p = g(p)

3) Khi do p la nghiem cu f(x) = 0

### Thuat toan

1) set i = 1
2) while i < N do step 3-6
3) p = g(p_0)
4) if abs (p-p0) < tol, output p, stop
5) i = i + 1
6) p_0 = p (update p0)
7) Output


```python
#BT x^3+4x^2-10 = 0


def diemcodinh_1(x):
    y =  np.sqrt(10-x**3)/2
    return y

def diemcodinh_2(x):
    y =  x - (x**3+4*x**2-10)
    return y

def diemcodinh_3(x):
    y =  np.sqrt(10/x -4*x)
    return y

def diemcodinh_4(x):
    y =  np.sqrt(10/(4+x))
    return y

def diemcodinh_5(x):
    y =  x - (x**3+4*x**2-10)/(3*x**2+8*x)
    return y
    
def diemcodinh_file(tol,p0,N,filename,hamso):
    with open(filename, "w", encoding="utf-8") as file:
        file.write("Giai phuong trinh f(x) = 0 bang phuong phap diem co dinh\n")
        file.write("#" * 70 + "\n")
        file.write(f"{'i':>3s}  {'p0':>12s}  {'f(p0)':>18s}\n")
        i = 1
        for i in range(0,N):
            p = hamso(p0)
            file.write(f"{i:3d}  {p0:12.10f}  {p:18.10f}\n")
            if abs(p - p0) < tol:
                break
            p0 = p
            if i == N - 1:
                file.write("!" * 70 + "\n" +"!" * 70 + "\n")
                file.write('\n Root NOT found after Nmax interation') 

tol = 1e-9
p0 = 1
N = 100
diemcodinh_file (tol,p0,N,'diemcodinh_1.dat',diemcodinh_1)
#diemcodinh_file (tol,p0,N,'diemcodinh_2.dat',diemcodinh_2)
#diemcodinh_file (tol,p0,N,'diemcodinh_3.dat',diemcodinh_3)
diemcodinh_file (tol,p0,N,'diemcodinh_4.dat',diemcodinh_4)
diemcodinh_file (tol,p0,N,'diemcodinh_5.dat',diemcodinh_5)
```

## Phuong phap Newton-Raphson

Chon ham g(x) = x-f(x)/f'(x)

p = p0 - f(p0)/f'(p0)

Khac voi p = f(p0)


```python
def newton_raphson(tol,p0,N,filename,hamso,daoham):
    with open(filename, "w", encoding="utf-8") as file:
        file.write("Giai phuong trinh f(x) = 0 bang phuong phap Newton Raphson\n")
        file.write("#" * 70 + "\n")
        file.write(f"{'i':>3s}  {'p':>12s}  {'f(p)':>18s}\n")
        i = 1
        for i in range(0,N):
            p = p0 - hamso(p0)/daoham(p0)
            file.write(f"{i:3d}  {p:12.10f}  {hamso(p):18.10f}\n")
            if abs(p - p0) < tol:
                break
            p0 = p
            if i == N - 1:
                file.write("!" * 70 + "\n" +"!" * 70 + "\n")
                file.write('\n Root NOT found after Nmax interation') 

tol = 1e-10
p0 = 1
N = 100

def hamso_newton(x):
    y =  np.exp(x) - 4*x - 5
    return y

def diff_hamso_newton(x):
    y =  np.exp(x) - 4
    return y

newton_raphson (tol,p0,N,'newton_raphson.dat',hamso_newton,diff_hamso_newton)
```

## Phuong phap Secant 
### Khac voi Newton-Rapson la phai tu lay dao ham, thi cai nay dung 

f'(p_n-1) = [f(p_n-2)-f(p_n-1)]/(p_n-2-p_n-1)

```python
def phuongphap_secant(tol,p0,p1,N,filename,hamso):
    with open(filename, "w", encoding="utf-8") as file:
        file.write("Giai phuong trinh f(x) = 0 bang phuong phap Secant\n")
        file.write("#" * 70 + "\n")
        file.write(f"{'i':>5s}  {'p0':>20s}  {'f(p0)':>20s}\n")
        i = 1
        for i in range(0,N):
            p = p1 - (hamso(p1)*(p1-p0))/(hamso(p1)-hamso(p0))
            file.write(f"{i:5d}  {p0:20.10f}  {hamso(p):20.10f}\n")
            if abs(p - p1) < tol:
                break
            p0 = p1
            p1 = p
            if i == N - 1:
                file.write("!" * 70 + "\n" +"!" * 70 + "\n")
                file.write('\n Root NOT found after Nmax interation') 

tol = 1e-10
p0 = 5
p1 = 2
N = 100

def hamso_newton(x):
    y =  np.exp(x) - 4*x - 5
    return y

phuongphap_secant(tol,p0,p1,N,'secant.dat',hamso_newton)
```
