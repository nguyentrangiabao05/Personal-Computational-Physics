## Phuong phap lap

Giai phuong trinh tuyen tinh n x n (DSTT)

Ax = b

Cho x(0) ban dau -> x(k) cho hoi tu

xi = 1/aii * (sigma (j=1, i khac j) -aij xj + bi), i = 1,2,...,n

Ben trai la dang muon tinh
Ben phai gia su nhu la da biet roi

xi (lan k) = 1/aii * (sigma (j=1, i khac j) -aij )

```python
import numpy as np

k = 1
N_max = 1000

A = [[10, -1, 2, 0], 
    [-1, 11, -1, 3], 
    [2, -1, 10, -1],
    [0, 3, -1, 8]]

B = [6, 25, -11, 15]

x_0 = [0, 0, 0, 0]

n = len(B)

epsilon = 1e-3

k = 0
err_mang = n * [1]
err = 1
x_new = n * [0]
x_old = n * [0]

x_old[:] = x_0[:]
```

```python

for q in range(N_max):
    if err > epsilon:
        for i in range(n):
            tong = 0
            for j in range(n):
                if i != j:
                    tong = tong + A[i][j] * x_old[j]
            x_new[i] = 1 / A[i][i] * (-tong + B[i])
            err_mang[i] = np.abs(x_new[i] - x_old[i])

        err = np.max(err_mang)

        x_old[:] = x_new[:]

        k = k + 1
    else:
        break

print("x =", x_new)
print("k =", k)
print("err =", err)
```

## Gauss Seider 

```python
for q in range(N_max):
    if err > epsilon:
        for i in range(n):
            tong1 = tong2 = 0
            for j in range(i):
                tong1 = tong1 + A[i][j] * x_new[j]
            for j in range(i+1,n):
                tong2 = tong2 + A[i][j] * x_old[j]
            
            x_new[i] = 1 / A[i][i] * (-tong1 -tong2 + B[i])
            
            err_mang[i] = np.abs(x_new[i] - x_old[i])

        err = np.max(err_mang)

        x_old[:] = x_new[:]

        k = k + 1
    else:
        break

print("x =", x_new)
print("k =", k)
print("err =", err)
```
