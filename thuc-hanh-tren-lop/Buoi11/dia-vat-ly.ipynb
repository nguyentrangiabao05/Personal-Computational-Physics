```python
import numpy as np
```

```python
def dieukienbien():
    global rho, A, N_z, N_t, z_max, L, t_max, T_mean, P, kappa, cp, Delta_T
    dz = z_max / (N_z -1)
    dt = t_max / (N_t -1)
    t = np.linspace(0, t_max, N_t)
    z = np.linspace(0, z_max, N_z)
    eta = kappa * dt /dz**2
    if eta > 0.5:
        print("Dieu kien bien khong duoc thoan, hay tang N_z hoac N_t")
        exit()
    index_L = int(L/dz)
    T = np.zeros((N_z, N_t))
    T[0,:] = T_mean + Delta_T * np.sin(2*np.pi*t[:]/P)
    T[index_L,:] = T_mean + 0.03 * L
    T[-1,:] = T_mean + 0.03 * z_max

    return T, z, t, dz, dt, eta, index_L
```

```python
def tinhtoan():
    T, z, t, dz, dt, eta, index_L = dieukienbien()
    for n in range(0, N_t-1): # Tranh dieu kien bien ra, luc dau la n+1 nen tinh tai index 1
        for j in range(1, N_z-1): # Tranh dieu kien bien ra
            if j == index_L:
                T[j,n+1] = T_mean + 0.03 * L #tai moi thoi diem t luon bang nhau, nen phai gan theo thoi gian (n+1)
                continue
            T[j,n+1] = T[j,n] + eta * (T[j+1,n] - 2*T[j,n] + T[j-1,n]) + dt * P / (rho * cp)
    return T
```
