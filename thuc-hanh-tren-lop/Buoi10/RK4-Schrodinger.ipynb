```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def create_grid(x_min=-10, x_max=10, dx=0.1, t_min=0, t_max=30, dt=0.01):
    x = np.arange(x_min, x_max, dx)
    t = np.arange(t_min, t_max + dt, dt)
    return x, t


def initial_wavefunction(x, x0=5, sigma=0.5):
    psi0 = np.exp(-(x - x0)**2 / (4 * sigma))
    return psi0.astype(complex)


def potential(x):
    V = np.zeros(len(x))
    return V


def build_matrix(V, dx):
    N = len(V)

    g = (-5j / (4 * dx**2) - 1) * V
    a = 1j / (24 * dx**2)

    diag = np.diag(g)

    off_diag1 = (
        np.diag([16 * a] * (N - 1), k=1)
        + np.diag([16 * a] * (N - 1), k=-1)
    )

    off_diag2 = (
        np.diag([-a] * (N - 2), k=2)
        + np.diag([-a] * (N - 2), k=-2)
    )

    M = diag + off_diag1 + off_diag2
    return M


def rk4_step(psi, M, dt):
    k1 = M @ psi
    k2 = M @ (psi + 0.5 * dt * k1)
    k3 = M @ (psi + 0.5 * dt * k2)
    k4 = M @ (psi + dt * k3)

    psi_next = psi + (dt / 6) * (k1 + 2*k2 + 2*k3 + k4)
    return psi_next


def solve_time_evolution(psi0, M, t, dt):
    Nt = len(t)
    N = len(psi0)

    y = np.zeros((Nt, N), dtype=complex)
    y[0] = psi0

    for n in range(Nt - 1):
        y[n + 1] = rk4_step(y[n], M, dt)

    return y


dx = 0.1
dt = 0.01

x, t = create_grid(
    x_min=-10,
    x_max=10,
    dx=dx,
    t_min=0,
    t_max=30,
    dt=dt
)

V = potential(x)
psi0 = initial_wavefunction(x, x0=5, sigma=0.5)

M = build_matrix(V, dx)

y = solve_time_evolution(psi0, M, t, dt)
```
