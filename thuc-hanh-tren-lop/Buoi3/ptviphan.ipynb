## Ordinary Differential Equation

### Euler

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def phuongtrinh1(x,y):
    return x

def euler(a,b,N,alpha,phuongtrinh,filename):
    h = (b-a)/N
    x = []
    y = []

    x.append(a)
    y.append(alpha)

    for i in range(1, N+1):
        y.append(y[i-1] + h * phuongtrinh(x[i-1], y[i-1]))
        x.append(a + i*h)

    with open(filename, "w", encoding="utf-8") as file:
        file.write("# Giai phuong trinh vi phan\n")
        file.write("#" * 50 + "\n")
        file.write(f"{'i':>6s}  {'x':>12s}  {'y':>12s}\n")

        for i in range(N+1):
            file.write(f"{i:6d}  {x[i]:12.9f}  {y[i]:12.9f}\n")
    return x, y

x_euler, y_euler = euler(0,10,100,0,phuongtrinh1,"Euler_PT1.txt")
```

```python
def phuongtrinh2(x,y):
    f = y - x**2 +1
    return f

x_Euler, y_Euler = euler(0,10,200,0.5,phuongtrinh2, "Euler_PT2.txt")
```

### Runge - Kutta methods, midpoint (RK2)

```python
def RK2_midpoint(a,b,N,alpha,phuongtrinh,filename):
    h = (b-a)/N
    x = []
    y = []

    x.append(a)
    y.append(alpha)

    for i in range(1, N+1):
        y.append(y[i-1] + h * phuongtrinh(x[i-1]+h/2, y[i-1]+h/2*phuongtrinh(x[i-1], y[i-1])))
        x.append(a + i*h)

    with open(filename, "w", encoding="utf-8") as file:
        file.write("# Giai phuong trinh vi phan\n")
        file.write("#" * 50 + "\n")
        file.write(f"{'i':>6s}  {'x':>12s}  {'y':>12s}\n")

        for i in range(N+1):
            file.write(f"{i:6d}  {x[i]:12.9f}  {y[i]:12.9f}\n")
    return x, y

def Euler_modified(a,b,N,alpha,phuongtrinh,filename):
    h = (b-a)/N
    x = []
    y = []

    x.append(a)
    y.append(alpha)

    for i in range(1, N+1):
        x.append(a + i*h)
        y.append(y[i-1] + h/2 * (phuongtrinh(x[i-1], y[i-1]) + phuongtrinh(x[i], y[i-1]+h*phuongtrinh(x[i-1],y[i-1]))))

    with open(filename, "w", encoding="utf-8") as file:
        file.write("# Giai phuong trinh vi phan\n")
        file.write("#" * 50 + "\n")
        file.write(f"{'i':>6s}  {'x':>12s}  {'y':>12s}\n")

        for i in range(N+1):
            file.write(f"{i:6d}  {x[i]:12.9f}  {y[i]:12.9f}\n")
    return x, y

x_RK2, y_RK2 = RK2_midpoint(0,10,200,0.5,phuongtrinh2, "RK2_PT2.txt")
x_Euler_mod, y_Euler_mod = Euler_modified(0,10,200,0.5,phuongtrinh2, "Euler_mod_PT2.txt")
```

### RK3

```python
def RK3(a,b,N,alpha,phuongtrinh,filename):
    h = (b-a)/N
    x = []
    y = []

    x.append(a)
    y.append(alpha)

    for i in range(1, N+1):
        x.append(a + i*h)
        k1 = h * phuongtrinh(x[i-1],y[i-1])
        k2 = h * phuongtrinh(x[i-1]+h/3,y[i-1]+1/3*k1)
        k3 = h * phuongtrinh(x[i-1]+2*h/3,y[i-1]+2/3*k2)
        y.append(y[i-1]+1/4*(k1+3*k3))

    with open(filename, "w", encoding="utf-8") as file:
        file.write("# Giai phuong trinh vi phan\n")
        file.write("#" * 50 + "\n")
        file.write(f"{'i':>6s}  {'x':>12s}  {'y':>12s}\n")

        for i in range(N+1):
            file.write(f"{i:6d}  {x[i]:12.9f}  {y[i]:12.9f}\n")
    return x, y

x_RK3, y_RK3  = RK3(0,10,200,0.5,phuongtrinh2, "RK3_PT2.txt")
```

### RK4

```python
def RK4(a, b, N, alpha, phuongtrinh, filename):
    h = (b - a) / N
    x = []
    y = []

    x.append(a)
    y.append(alpha)

    for i in range(1, N + 1):
        x.append(a + i*h)

        k1 = h * phuongtrinh(x[i-1], y[i-1])
        k2 = h * phuongtrinh(x[i-1] + h/2, y[i-1] + k1/2)
        k3 = h * phuongtrinh(x[i-1] + h/2, y[i-1] + k2/2)
        k4 = h * phuongtrinh(x[i-1] + h,   y[i-1] + k3)

        y.append(y[i-1] + (k1 + 2*k2 + 2*k3 + k4)/6)

    with open(filename, "w", encoding="utf-8") as file:
        file.write("# Giai phuong trinh vi phan\n")
        file.write("#" * 50 + "\n")
        file.write(f"{'i':>6s}  {'x':>12s}  {'y':>12s}\n")

        for i in range(N + 1):
            file.write(f"{i:6d}  {x[i]:12.9f}  {y[i]:12.9f}\n")
    return x, y

x_RK4, y_RK4 = RK4(0,10,200,0.5,phuongtrinh2, "RK4_PT2.txt")
```

### Euler method for SoDE

```python
def machdien(I1, I2):
    f1 = -4*I1 + 3*I2 + 6
    f2 = -2.4*I1 + 1.6*I2 + 3.6
    return f1, f2

def euler_nhieu(a, b, N, alpha, phuongtrinh, filename):
    h = (b - a) / N

    x = [a]
    y = [alpha[:] ]   # alpha = [I1_0, I2_0]

    for i in range(N):
        I1, I2 = y[i]
        f1, f2 = phuongtrinh(I1, I2)

        I1_moi = I1 + h * f1
        I2_moi = I2 + h * f2

        y.append([I1_moi, I2_moi])
        x.append(a + (i + 1) * h)

    with open(filename, "w", encoding="utf-8") as file:
        file.write("# Giai he phuong trinh vi phan bang phuong phap Euler\n")
        file.write("#" * 60 + "\n")
        file.write(f"{'i':>6s} {'t':>12s} {'I1':>12s} {'I2':>12s}\n")

        for i in range(N + 1):
            file.write(f"{i:6d} {x[i]:12.9f} {y[i][0]:12.9f} {y[i][1]:12.9f}\n")

    return x, y

def Modified_euler_nhieu(a, b, N, alpha, phuongtrinh, filename):
    h = (b - a) / N

    x = [a]
    y = [alpha[:] ]   # alpha = [I1_0, I2_0]

    for i in range(N):
        I1, I2 = y[i]

        f1, f2 = phuongtrinh(I1, I2)

        I1_du_doan = I1 + h * f1
        I2_du_doan = I2 + h * f2

        f1_moi, f2_moi = phuongtrinh(I1_du_doan, I2_du_doan)

        I1_moi = I1 + h/2 * (f1 + f1_moi)
        I2_moi = I2 + h/2 * (f2 + f2_moi)

        y.append([I1_moi, I2_moi])
        x.append(a + (i + 1) * h)

    with open(filename, "w", encoding="utf-8") as file:
        file.write("# Giai he phuong trinh vi phan bang phuong phap Euler\n")
        file.write("#" * 60 + "\n")
        file.write(f"{'i':>6s} {'t':>12s} {'I1':>12s} {'I2':>12s}\n")

        for i in range(N + 1):
            file.write(f"{i:6d} {x[i]:12.9f} {y[i][0]:12.9f} {y[i][1]:12.9f}\n")

    return x, y

alpha = [0,0]
t, I_euler = euler_nhieu(0, 10, 100, alpha, machdien, "machdien_euler")
t, I_eulercaitien = Modified_euler_nhieu(0, 10, 100, alpha, machdien, "machdien_euler_caitien")
```

```python
def euler_nhieu(a, b, N, alpha, phuongtrinh, filename):
    h = (b - a) / N
    m = len(alpha)

    x = [a]
    y = [alpha[:] ]

    for i in range(N):
        Y_hien_tai = y[i][:]
        F = phuongtrinh(Y_hien_tai)   # F = [f1, f2, ..., fm]
        
        Y_moi = []
        for j in range(m):
            Y_moi.append(Y_hien_tai[j] + h * F[j])

        y.append(Y_moi)
        x.append(a + (i + 1) * h)

    with open(filename, "w", encoding="utf-8") as file:
        file.write("# Giai he phuong trinh vi phan bang phuong phap Euler\n")
        file.write("#" * 80 + "\n")

        # header
        file.write(f"{'i':>6s} {'t':>12s}")
        for j in range(m):
            file.write(f" {'y'+str(j+1):>12s}")
        file.write("\n")

        # data
        for i in range(N + 1):
            file.write(f"{i:6d} {x[i]:12.9f}")
            for j in range(m):
                file.write(f" {y[i][j]:12.9f}")
            file.write("\n")

    return x, y
```

### Higher-order differential equations (HODE -> SoDE)

```python
import numpy as np

def phuongtrinh_daodong(theta, v):
    g = 9.8
    l = 0.3
    dthetadt = v
    dvdt = -g/l * np.sin(theta)
    return dthetadt, dvdt


def RK4(a, b, N, alpha, phuongtrinh, filename):
    h = (b - a) / N
    x = []
    y = []

    x.append(a)
    y.append(alpha[:])   # alpha = [theta0, v0]

    for i in range(N):
        theta, v = y[i]

        k1_theta, k1_v = phuongtrinh(theta, v)
        k1_theta *= h
        k1_v *= h

        k2_theta, k2_v = phuongtrinh(theta + k1_theta/2, v + k1_v/2)
        k2_theta *= h
        k2_v *= h

        k3_theta, k3_v = phuongtrinh(theta + k2_theta/2, v + k2_v/2)
        k3_theta *= h
        k3_v *= h

        k4_theta, k4_v = phuongtrinh(theta + k3_theta, v + k3_v)
        k4_theta *= h
        k4_v *= h

        theta_moi = theta + (k1_theta + 2*k2_theta + 2*k3_theta + k4_theta)/6
        v_moi     = v     + (k1_v     + 2*k2_v     + 2*k3_v     + k4_v    )/6

        y.append([theta_moi, v_moi])
        x.append(a + (i + 1)*h)

    with open(filename, "w", encoding="utf-8") as file:
        file.write("# Giai he phuong trinh vi phan bang RK4\n")
        file.write("#" * 60 + "\n")
        file.write(f"{'i':>6s} {'t':>12s} {'theta':>12s} {'v':>12s}\n")

        for i in range(N + 1):
            file.write(f"{i:6d} {x[i]:12.9f} {y[i][0]:12.9f} {y[i][1]:12.9f}\n")

    return x, y

alpha = [np.deg2rad(30), 0]
x_daodong, y_daodong = RK4(0,10, 1000, alpha, phuongtrinh_daodong, "Daodongdieuhoa.txt")
```
