```python
## Phuong trinh Schrodinger phu thuoc thoi gian
import numpy as np
import matplotlib.pyplot as plt
plt.style.use('sci.mplstyle')
```

```python
delta_x = 0.02
delta_t = 1/2 * delta_x **2
k_0 = 17 * np.pi

x_max = 15
x_min = 0
t_max = 30
t_min = 0

sigma_0 = 0.5
hbar = 1
m = 1
```

```python
N_x = np.int64((x_max-x_min)/delta_x) +1
N_time =  np.int64((t_max-t_min)/delta_t) +1

def khoi_tao_ham_song_ban_dau():
    global delta_x, delta_t, k_0, x_max, x_min, t_max, t_min, sigma_0, hbar, m, N_x, N_time
    time_arr = np.linspace(t_min,t_max, N_time)
    x_arr = np.linspace(x_min, x_max, N_x)
    psi_zero = np.zeros((N_x,N_time), dtype=np.complex128)

    psi_zero[:,0] = np.exp(-1/2 * ((x_arr-5)/sigma_0)**2) * np.exp(1j*k_0*x_arr)
    psi_zero[-1,:] = 0
    psi_zero[0,:] = 0

    return time_arr, x_arr, psi_zero
```

```python
def khoi_tao_the_nang_ban_dau():
    global time_arr, x_arr, N_time, N_x
    V_0 = np.zeros((N_x,N_time))
    for i in range(len(x_arr)):
        if x_arr[i] >= 0 and x_arr[i] <= 15:
            V_0[i][:] = 0
        else:
            V_0[i][:] = 999

    return V_0
```

```python
def giai_ham_song(psi_zero, V_0):
    global time_arr, x_arr, delta_x, delta_t, N_time, N_x,hbar,m
    g_j = 1 - 1j*delta_t/hbar * V_0 - 1j*hbar*delta_t/(m*delta_x**2)
    a = 1j * hbar * delta_t / (2*m*delta_x**2)
    psi = psi_zero
    for q in range(0,N_time-1):
        for j in range(1,N_x-1):
            psi[j,q+1] = g_j[j,q] * psi [j,q] + a*(psi [j+1,q]+psi [j-1,q]) # q bang 0 thi se tinh cho q = 1

    psi_zero[-1,:] = 0
    psi_zero[0,:] = 0
    return psi
```

```python
time_arr, x_arr, psi_zero = khoi_tao_ham_song_ban_dau()
V_0 = khoi_tao_the_nang_ban_dau()

psi = giai_ham_song(psi_zero, V_0)
```
