```python
import numpy as np
```

```python
def ham_Ising(m, T):
    return m - np.tanh(m / T)

def df_ham_Ising(m, T):
    giatri_tanh = np.tanh(m / T)
    return 1 - (1 / T) * (1 - giatri_tanh**2)

def chiadoikhoangcach(a, b, tol, N, hamso, T):
    fa = hamso(a, T)
    fb = hamso(b, T)
    if fa * fb > 0:
        return 0.0
    
    for i in range(0, N):
        c = (a + b) / 2
        fc = hamso(c, T)
        if fa * fc < 0:
            b = c
        else:
            a = c
        if abs(fc) < tol:
            break
    return c

def newton_raphson(m0, tol, N, hamso, d_hamso, T):
    m = m0
    for i in range(0, N):
        f_val = hamso(m, T)
        df_val = d_hamso(m, T)
        if abs(df_val) < 1e-12:
            break
        m_next = m - f_val / df_val
        f_next = hamso(m_next, T)
        if abs(m_next - m) < tol or abs(f_next) < tol:
            m = m_next
            break
        m = m_next
    return m
```

```python
T_arr = np.linspace(0.01, 1.5, 300)
m_bisection = []
m_newton = []

tol = 1e-6
N = 500

for i, T in enumerate(T_arr):
    if T < 1.0: #Chon 2 vung nghiem de tranh nghiem tam thuong
        a1 = 1e-8
        b1 = 1.5
    else:
        a1 = -0.5
        b1 = 0.5
        
    m_bis = chiadoikhoangcach(a1, b1, tol, N, ham_Ising, T)
    m_bisection.append(m_bis)
    
    m0 = 1.0
    m_newt = newton_raphson(m0, tol, N, ham_Ising, df_ham_Ising, T)
    m_newton.append(m_newt)

output_filename = "Nghiem_ISING.txt"
with open(output_filename, "w", encoding="utf-8") as f:
    f.write(f"#{'i':>5s}  {'T':>12s}  {'Bisection (m)':>18s}  {'Newton (m)':>18s}\n")
    f.write("#" * 60 + "\n")
    for idx, T in enumerate(T_arr):
        f.write(f"{idx+1:5d}  {T:12.4f}  {m_bisection[idx]:18.10f}  {m_newton[idx]:18.10f}\n")
```
