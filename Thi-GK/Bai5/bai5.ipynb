```python
import numpy as np

sigma = 10
beta = 8/3
rho = 28

t_batdau = 0
t_ketthuc = 50        
dt = 0.005
N = int((t_ketthuc - t_batdau) / dt)

alpha = np.array([1.0, 1.0, 1.0]) 
```

```python
def he_phuongtrinh(t, state):
    x, y, z = state
    dxdt = sigma * (y - x)
    dydt = x * (rho - z) - y
    dzdt = x * y - beta * z
    return np.array([dxdt, dydt, dzdt])

def RK4(a, b, N, alpha, phuongtrinh, filename):
    h = (b - a) / N
    t_giatri = []
    trangthai_giatri = []

    t_giatri.append(a)
    trangthai_giatri.append(alpha)

    for i in range(1, N + 1):
        t_giatri.append(a + i * h)

        k1 = h * phuongtrinh(t_giatri[i-1], trangthai_giatri[i-1])
        k2 = h * phuongtrinh(t_giatri[i-1] + h/2, trangthai_giatri[i-1] + k1/2)
        k3 = h * phuongtrinh(t_giatri[i-1] + h/2, trangthai_giatri[i-1] + k2/2)
        k4 = h * phuongtrinh(t_giatri[i-1] + h, trangthai_giatri[i-1] + k3)

        trangthai_giatri.append(trangthai_giatri[i-1] + (k1 + 2*k2 + 2*k3 + k4) / 6)

    with open(filename, "w", encoding="utf-8") as file:
        file.write("# Giai phuong trinh vi phan (He PT Lorenz)\n")
        file.write("#" * 70 + "\n")
        file.write(f"# {'i':>6s}  {'t':>12s}  {'x':>12s}  {'y':>12s}  {'z':>12s}\n")

        for i in range(N + 1):
            file.write(f" {i:6d}  {t_giatri[i]:12.5f}  {trangthai_giatri[i][0]:12.6f}  {trangthai_giatri[i][1]:12.6f}  {trangthai_giatri[i][2]:12.6f}\n")
            
    return t_giatri, trangthai_giatri
```

```python
t_points, solution_points = RK4(t_batdau, t_ketthuc, N, alpha, he_phuongtrinh, "ket_qua_lorenz.txt")
```
