## Bai toan truyen nhiet

alpha^2 * d^2(u)/dx^2 = d(u)/dt

Chieu dai thanh l, 2 dau luon de 0 do C

Dieu kien bien: T(x,t=0)=100, va T(0,t)=T(l,t) = 0

## Dieu kien on dinh bai toan
alpha^2 k/h^2 <= 1/2

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def dieukienbien(N_time,N_x,x_t0,t_x0,t_xl):
    u = np.zeros((N_x,N_time),dtype=float)

    #bien dau la t, bien thu 2 la x
    #x_t0 la tai thoi diem ban dau T = 100

    for i in range(N_time):
        u[N_x-1][i] = (t_xl)

    for i in range(N_time):
        u[0][i] = (t_x0)

    for i in range(N_x):
        u[i][0] = (x_t0)

    return u
```

```python
def hamso(kappa,C, rho, N_x, N_time,l,t_max):
    h = l/(N_x-1)
    k = t_max/(N_time-1)
    alpha = np.sqrt(kappa/(C*rho))
    eta  = alpha ** 2 * k / h**2

    if eta <= 0.5:
        return eta,h,k
    else:
        print("Eta da lon hon 0.5, khong the thuc hien duoc, eta = ", eta)
        return False
```

```python
def tinhtoan(u, kappa, C, rho, N_x, N_time, l, t_max):
    result,h,k = hamso(kappa, C, rho, N_x, N_time, l, t_max)

    if isinstance(result, tuple):
        eta = result[0]         
    else:
        eta = result

    for j in range(N_time-1):
        for i in range(1, N_x-1):
            u[i][j+1] = (1 - 2*eta)*u[i][j] + eta*(u[i+1][j] + u[i-1][j])

    filename = "truyen_nhiet.dat"
    with open(filename, "w", encoding="utf-8") as file:
        file.write("# Bai toan truyen nhiet 1D\n")
        file.write("# Phuong phap sai phan huu han hien\n")
        file.write("#\n")
        file.write(f"# kappa  = {kappa}\n")
        file.write(f"# C      = {C}\n")
        file.write(f"# rho    = {rho}\n")
        file.write(f"# N_x    = {N_x}\n")
        file.write(f"# N_time = {N_time}\n")
        file.write(f"# l      = {l}\n")
        file.write(f"# t_max  = {t_max}\n")
        file.write(f"# eta    = {eta}\n")
        file.write("#\n")
        file.write("# Cot du lieu: j   i   t   x   u\n")

        for j in range(N_time):
            t = j * k
            for i in range(N_x):
                x = i * h
                file.write(f"{j:5d} {i:5d} {t:15.8e} {x:15.8e} {u[i][j]:15.8e}\n")
            file.write("\n")
    return u
```

```python
# Dieu kien bien

N_time = 6000
N_x = 100
x_t0 = 100 #Do C
t_x0 = 0 #Do C
t_xl = 0 #Do C

# Tham so bai toan
kappa  = 210
C = 900
rho = 2700
l = 1

t_max = 1800

u = dieukienbien(N_time,N_x,x_t0,t_x0,t_xl)
```

```python
solution  = tinhtoan(u, kappa,C, rho, N_x, N_time,l,t_max)
```

## Muc 20.3, p. 483 – 484 sach R Landau

```python
## Bai toan 2 thanh 0.25 cm

def dieukienbien_2thanh(N_time, N_x, T_left, T_right, l_bar):

    l_total = 2.0 * l_bar
    x = np.linspace(0.0, l_total, N_x)

    u = np.zeros((N_x, N_time), dtype=float)

    # Dieu kien dau: thanh trai 100 K, thanh phai 50 K
    for i in range(N_x):
        if x[i] <= l_bar:
            u[i, 0] = T_left
        else:
            u[i, 0] = T_right

    # Dieu kien bien: hai dau ngoai cung giu 0 K cho moi thoi diem
    u[0, :] = 0.0
    u[-1, :] = 0.0

    # de chac chan tai t=0 hai bien van la 0
    u[0, 0] = 0.0
    u[-1, 0] = 0.0

    return u, x
```

```python
def tinhtoan(u, kappa, C, rho, N_x, N_time, l, t_max):
    eta, h, k = hamso(kappa, C, rho, N_x, N_time, l, t_max)

    if eta > 0.5:
        raise ValueError(
            f"So do explicit khong on dinh: eta = {eta:.6f} > 0.5. "
            "Hay tang N_time hoac giam N_x."
        )

    for j in range(N_time - 1):
        for i in range(1, N_x - 1):
            u[i, j + 1] = (1 - 2*eta)*u[i, j] + eta*(u[i + 1, j] + u[i - 1, j])

    filename = "truyen_nhiet_2thanh.dat"
    with open(filename, "w", encoding="utf-8") as file:
        file.write("# Bai toan truyen nhiet 1D: hai thanh tiep xuc\n")
        file.write("# Phuong phap sai phan huu han hien\n")
        file.write("# Cot du lieu: j   i   t   x   u\n")
        file.write(f"# kappa  = {kappa}\n")
        file.write(f"# C      = {C}\n")
        file.write(f"# rho    = {rho}\n")
        file.write(f"# N_x    = {N_x}\n")
        file.write(f"# N_time = {N_time}\n")
        file.write(f"# l      = {l}\n")
        file.write(f"# t_max  = {t_max}\n")
        file.write(f"# eta    = {eta}\n")
        file.write("#\n")

        for j in range(N_time):
            t = j * k
            for i in range(N_x):
                x = i * h
                file.write(f"{j:5d} {i:5d} {t:15.8e} {x:15.8e} {u[i, j]:15.8e}\n")
            file.write("\n")

    return u
```

```python
N_time = 15000          # phai du lon de explicit on dinh
N_x = 101               # nen lay le de x = 0.25 nam dung tren 1 nut luoi

T_left = 100.0          # thanh trai
T_right = 50.0          # thanh phai

t_x0 = 0.0              # bien trai
t_xl = 0.0              # bien phai

kappa = 210.0
C = 900.0
rho = 2700.0

l_bar = 0.25
l_total = 2.0 * l_bar   # = 0.5 m
t_max = 1800.0
```

```python
u, x = dieukienbien_2thanh(N_time, N_x, T_left, T_right, l_bar)
u = tinhtoan(u, kappa, C, rho, N_x, N_time, l_total, t_max)
```
