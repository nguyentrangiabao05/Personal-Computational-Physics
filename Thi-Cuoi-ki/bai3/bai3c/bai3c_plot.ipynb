```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.colors as mcolors

plt.style.use('sci.mplstyle')
```

```python
def doc_file_ketqua(filename): #dung de doc file ket qua
    x, y, V, err = np.loadtxt(filename, unpack=True, comments="#")

    # Lay so diem luoi tu file

    # unique co nghia la so day x ma co so khac nhau
    x_unique = np.unique(x)
    y_unique = np.unique(y)

    # lay len se ra duoc so diem x va y khac nhau -> luoi
    nx = len(x_unique)
    ny = len(y_unique)

    print("File:", filename)
    print("nx =", nx)
    print("ny =", ny)
    print("So diem trong file =", len(x))
    print("nx * ny =", nx * ny)

    # Neu file luu theo for i roi luu for j
    X = x.reshape(nx, ny)
    Y = y.reshape(nx, ny)
    V2D = V.reshape(nx, ny)
    err2D = err.reshape(nx, ny)

    return x_unique, y_unique, X, Y, V2D, err2D
```

```python
x_unique_gs_bai3c_giongnhau, y_unique_gs_bai3c_giongnhau, X_gs_bai3c_giongnhau, Y_gs_bai3c_giongnhau, V2D_gs_bai3c_giongnhau, err2D_gs_bai3c_giongnhau =  doc_file_ketqua("bai3c_giongnhau_gs_result.txt")


plt.figure(figsize=(8, 7))
plt.imshow(V2D_gs_bai3c_giongnhau.T, origin='lower', cmap='coolwarm',extent=[0, 100, 0, 100], )
plt.title("Hieu dien the 2D (Python)")
plt.xlabel('x (cm)')
plt.ylabel('y (cm)')
cb = plt.colorbar(label='Voltage (V)', orientation='vertical', pad=0.02)

plt.savefig("Thetinhdien-bai3c_giongnhau-gs.png", dpi=300)
plt.show()
```

```python
fig = plt.figure(figsize=(9, 9))
ax = fig.add_subplot(111, projection='3d')

ax.plot_surface(X_gs_bai3c_giongnhau, Y_gs_bai3c_giongnhau, V2D_gs_bai3c_giongnhau, color='white', edgecolor='gray', linewidth=0.3)
ax.set_xlabel('x (cm)')
ax.set_ylabel('y (cm)')
ax.set_zlabel('V')
ax.set_title("3D tinh dien (Python)")
ax.view_init(elev=20, azim=20)
plt.savefig("Thetinhdien-bai3c-giongnhau-3D.png", dpi=300)
plt.show()
```
