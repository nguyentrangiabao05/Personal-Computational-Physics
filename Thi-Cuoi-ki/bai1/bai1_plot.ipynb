```python
import numpy as np
import matplotlib.pyplot as plt
plt.style.use('sci.mplstyle')
```

```python
R_c_MC, I_z_MC = np.loadtxt("KQ_MC_bai1.txt", comments="#", unpack=True)
R_c_Riemann, I_z_Riemann = np.loadtxt("KQ_Riemann_bai1.txt", comments="#", unpack=True)


plt.figure(figsize=(8,6))
plt.plot(R_c_MC, I_z_MC,      label="I_z MC", color="C0", lw = 3)
plt.plot(R_c_MC, I_z_Riemann, label="I_z Riemann", color="C1")


plt.xlabel(f"R_core/R")
plt.ylabel("Moment quan tinh (I_z)")
plt.title("Bai 1: Python")
plt.legend()
plt.xlim(0,1)
plt.ylim(6,25)
plt.savefig("so_sanh_bai1.png", dpi = 300)
plt.show()
```
