```python
import numpy as np
import matplotlib.pyplot as plt
plt.style.use('sci.mplstyle')
import time
```

```python
def hammatdo(x,y,z, R_c):
    global R_p, rho_c, rho_m #R_p la ban kinh hanh tinh, R_c la ban kinh loi, rho_c la mat do loi, rho_m la mat do lop phu
    ban_kinh = np.sqrt(x**2+y**2+z**2)
    if 0 <= ban_kinh <= R_c:
        rho = rho_c # Mat do cua sat
    elif R_c < ban_kinh <= R_p:
        rho = rho_m # Mat do cua da
    else:
        rho = 0 # ngoai qua cau thi bang 0
    return rho
```

```python
def tinhMC(N, R_c):
    global R_p, rho_c, rho_m
    thanhphantinhtong = 0
    for i in range(N):
        x = np.random.uniform(-R_p, R_p) # tao random trong hinh hop tu -R_p den R_p
        y = np.random.uniform(-R_p, R_p) # mang random nay bao phu het toan bo hinh cau
        z = np.random.uniform(-R_p, R_p)

        rho = hammatdo(x,y,z, R_c) #tinh mat do
        thanhphantinhtong = thanhphantinhtong + rho * (x**2+y**2) #moment theo truc z
        # khong can dieu kien if vi rho da cho = 0 neu vuot qua mien qua cau. 
        
    moment_quantinh = thanhphantinhtong * (2*R_p)**3 / N

    return moment_quantinh
```

```python
def ghifile(R_c, I_z,thoigian_chay_MC, filename):
    filename = f"{"KQ"}_{filename}.txt"
    with open(filename, "w", encoding="utf-8") as file:
        file.write(f"# Tinh moment quan tinh bang MC va Riemann \n")
        file.write(f"# Thoi gian chay la: {thoigian_chay_MC:.12f} s\n")
        file.write("#\n")
        file.write(f"# {'R_c':>35s} {'Moment quan tinh(I_z)':>35s}\n")

        for R_c_step in range(len(R_c)):
                file.write(f"  {R_c[R_c_step]:35.8e} {I_z[R_c_step]:35.8e}\n")

# file ghi ket qua R_c theo I_z cung voi thoi gian chay
```

```python
N = 10**4 # so diem random

R_p = 1.0
rho_c = 13
rho_m = 4.5

# Tham so de bai
```

```python
R_c = np.linspace(0,R_p,50) #Tao mang cho R_C can khao sat

moment_quantinh_R_C = np.zeros(len(R_c)) # tao mang cho moment quan tinh
```

```python
thoigian_batdau_MC = time.perf_counter() # bat dau dem thoi gian chay

for i in range(len(R_c)):
    moment_quantinh_R_C[i] = tinhMC(N, R_c[i])

thoigian_kethuc_MC = time.perf_counter() # dem thoi gian ket thuc

thoigian_chay_MC = thoigian_kethuc_MC - thoigian_batdau_MC #tinh thoi gian chay
print(f"Thoi gian chay bang Monte Carlo la (s): ", thoigian_chay_MC)

ghifile(R_c, moment_quantinh_R_C,thoigian_chay_MC, "MC_bai1") #ghi ra file
```

# Tong Riemann

```python
N_Riemann = 50 #so diem chia ra trong tong Riemann
```

```python
def tinh_Riemann(N_Riemann, R_c):
    global R_p, rho_c, rho_m
    moment_quantinh = 0
    x_arr = np.linspace(-R_p, R_p, N_Riemann) #tao ra mang voi N_riemann gia chi diem chia
    y_arr = np.linspace(-R_p, R_p, N_Riemann)
    z_arr = np.linspace(-R_p, R_p, N_Riemann)

    dx = x_arr[1]-x_arr[0] # lay dx de tinh vi phan the tich
    dy = y_arr[1]-y_arr[0]
    dz = z_arr[1]-z_arr[0]

    dV = dx * dy * dz # tinh vi phan the tich 

    for i in range(len(x_arr)):
        for j in range(len(y_arr)):
            for k in range(len(z_arr)):
                rho = hammatdo(x_arr[i],y_arr[j],z_arr[k], R_c) #tinh mat do #
                # mat do da trong bang 0 o ngoai nen khong can dung if
                moment_quantinh = moment_quantinh + rho * (x_arr[i]**2+y_arr[j]**2) * dV #moment theo truc z
                # vi phan moment quan tinh nhan rho * dV * (x^2+y^2)
    return moment_quantinh
```

```python
R_c = np.linspace(0,R_p,50) #Tao mang cho R_C can khao sat

moment_quantinh_R_C_Riemann = np.zeros(len(R_c)) # tao mang cho moment quan tinh

thoigian_batdau_tong_Riemann = time.perf_counter() #thoi gian bat dau

for i in range(len(R_c)):
    moment_quantinh_R_C_Riemann[i] = tinh_Riemann(N_Riemann, R_c[i]) # chay theo diem can khao sat

thoigian_kethuc_tong_Riemann = time.perf_counter() #thoi gian ket thuc

thoigian_chay_tong_Riemann = thoigian_kethuc_tong_Riemann - thoigian_batdau_tong_Riemann #thoi gian chay
print(f"Thoi gian chay bang Riemann la (s): ", thoigian_chay_tong_Riemann)

ghifile(R_c, moment_quantinh_R_C_Riemann,thoigian_chay_tong_Riemann, "Riemann_bai1") #ghi ra file de so sanh
```
