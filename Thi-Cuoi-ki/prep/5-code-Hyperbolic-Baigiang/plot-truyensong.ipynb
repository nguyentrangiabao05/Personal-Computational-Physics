```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.colors as mcolors
from matplotlib.lines import Line2D

plt.style.use('sci.mplstyle')
```

```python
def ham_doc_file(filename):
    data = np.loadtxt(filename, comments="#")

    t_raw = data[:, 2]
    x_raw = data[:, 3]
    u_raw = data[:, 4]

    x = np.unique(x_raw)
    t = np.unique(t_raw)

    N_x = len(x)
    N_time = len(t)

    print("File:", filename)
    print("N_x =", N_x)
    print("N_time =", N_time)
    print("So diem =", len(u_raw))
    print("N_x * N_time =", N_x * N_time)
    print("\n\n")

    if N_x * N_time != len(u_raw):
        raise ValueError("File khong phai luoi day du theo x va t")

    # file ghi theo:
    # for j in range(N_time):
    #     for i in range(N_x):
    temp = u_raw.reshape(N_time, N_x)

    return x, t, temp
```

```python
x_bai1_friction, t_bai1_friction, u_bai1_friction = ham_doc_file("bai1_friction_beta_0.904.txt")

x_bai2_friction, t_bai2_friction, u_bai2_friction = ham_doc_file("bai2_friction_beta_0.904.txt")

x_bai1_no_friction, t_bai1_no_friction, u_bai1_no_friction = ham_doc_file("bai1_no_friction_beta_0.904.txt")

x_bai2_no_friction, t_bai2_no_friction, u_bai2_no_friction = ham_doc_file("bai2_no_friction_beta_0.904.txt")
```

```python
fig = plt.figure(figsize=(9, 9))
ax = fig.add_subplot(111, projection='3d')

X_2thanh, T_2thanh = np.meshgrid(x_bai1_no_friction, t_bai1_no_friction)

ax.plot_surface(X_2thanh, T_2thanh, u_bai1_no_friction, color='white', edgecolor='gray', linewidth=0.3)
ax.set_xlabel('x')
ax.set_ylabel('t')
ax.set_zlabel('Temp')
ax.view_init(elev=20, azim=20)
plt.savefig("bai1_no_friction_beta_0.904.png", dpi=300)
plt.show()
```
