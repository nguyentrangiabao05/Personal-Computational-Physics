```python
import numpy as np
import matplotlib.pyplot as plt
plt.style.use('sci.mplstyle')
import time
```

```python
def hammatdo(x,y,z):
    global R_p, R_c, rho_c, rho_m #R_p la ban kinh hanh tinh, R_c la ban kinh loi, rho_c la mat do loi, rho_m la mat do lop phu
    ban_kinh = np.sqrt(x**2+y**2+z**2)
    if 0 <= ban_kinh <= R_c:
        rho = rho_c
    elif R_c < ban_kinh <= R_p:
        rho = rho_m
    else:
        rho = 0
    return rho

def tinhMC(N):
    global R_p, R_c, rho_c, rho_m
    thanhphantinhtong = 0
    for i in range(N):
        x = np.random.uniform(-R_p, R_p)
        y = np.random.uniform(-R_p, R_p)
        z = np.random.uniform(-R_p, R_p)

        rho = hammatdo(x,y,z)
        thanhphantinhtong = thanhphantinhtong + rho * (x**2+y**2)
        
    moment_quantinh = thanhphantinhtong * (2*R_p)**3 / N

    return moment_quantinh
```

```python
N = 1000000
R_p = 6.4*1e6
R_c = 3.5*1e6
rho_c = 13000
rho_m = 4500
```

```python
moment_quantinh = tinhMC(N)
```
