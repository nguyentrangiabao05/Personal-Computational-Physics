```python
import numpy as np
import matplotlib.pyplot as plt
plt.style.use('sci.mplstyle')
import time
```

```python
N = 10**6

R = 1

a = -R
b = R
```

```python
# x**2 + y**2 <= R**2

def tru_A(x,y,z):
    tru_A = x**2+z**2
    return tru_A

def tru_B(x,y,z):
    tru_B = y**2 + z**2
    return tru_B

def tru_C(x,y,z):
    tru_C = x**2 + y**2
    return tru_C
```

```python
def MC_int_giao_2_tru(N, tru_A, tru_B, a, b):
    global R
    N_in = 0
    for i in range(N):
        x = np.random.uniform(a,b)
        y = np.random.uniform(a,b)
        z = np.random.uniform(a,b)
        if tru_A(x,y,z) <= R**2 and tru_B(x,y,z) <= R**2:
            N_in = N_in +1
    return N_in
```

```python
def Riemann_giao_2_tru(tru_A, tru_B, a, b, N_riemann):
    global R
    V = 0
    x = np.linspace(a,b,N_riemann)
    y = np.linspace(a,b,N_riemann)
    z = np.linspace(a,b,N_riemann)
    dx = x[1]-x[0]
    dy = y[1]-y[0]
    dz = z[1]-z[0]

    dV = dx * dy * dz
    for i in range(len(x)):
        for j in range(len(y)):
            for k in range(len(z)):
                if tru_A(x[i],y[j],z[k]) <= R**2 and tru_B(x[i],y[j],z[k]) <= R**2:
                    V = V + dV * 1
                else:
                    V = V
    return V
```

```python
N_in_2tru = MC_int_giao_2_tru(N, tru_A, tru_B, a, b)

V_0 = np.abs(a-b)**3

V_2tru = V_0 * N_in_2tru/N
```

```python
print("The tich phan giao nhau MC = ", f"{V_2tru:.6f}")
print("The tich giai tich = ", f"{16/3 * R**3:.6f}")
```

```python
N_riemann = 200
V_riemann_2tru = Riemann_giao_2_tru(tru_A, tru_B, a, b, N_riemann)
```

```python
print("The tich phan giao nhau Riemann = ", f"{V_riemann_2tru:.6f}")
```

# 3 tru giao nhau

```python
def MC_int_giao_3_tru(N, tru_A, tru_B, a, b):
    global R
    N_in = 0
    for i in range(N):
        x = np.random.uniform(a,b)
        y = np.random.uniform(a,b)
        z = np.random.uniform(a,b)
        if tru_A(x,y,z) <= R**2 and tru_B(x,y,z) <= R**2 and tru_C(x,y,z) <= R**2:
            N_in = N_in +1
    return N_in

N_in_3_tru = MC_int_giao_3_tru(N, tru_A, tru_B, a, b)

V_0 = np.abs(a-b)**3

V_3tru = V_0 * N_in_3_tru/N
```

```python
def Riemann_giao_3_tru(tru_A, tru_B, a, b, N_riemann):
    global R
    V = 0
    x = np.linspace(a,b,N_riemann)
    y = np.linspace(a,b,N_riemann)
    z = np.linspace(a,b,N_riemann)
    dx = x[1]-x[0]
    dy = y[1]-y[0]
    dz = z[1]-z[0]

    dV = dx * dy * dz
    for i in range(len(x)):
        for j in range(len(y)):
            for k in range(len(z)):
                if tru_A(x[i],y[j],z[k]) <= R**2 and tru_B(x[i],y[j],z[k]) <= R**2 and tru_C(x[i],y[j],z[k]) <= R**2:
                    V = V + dV * 1
                else:
                    V = V
    return V
```

```python
print("The tich phan giao nhau MC = ", f"{V_3tru:.6f}")
print("The tich giai tich = ", f"{(8*(2-np.sqrt(2)) * R**3):.6f}")
```

```python
V_riemann_3tru = Riemann_giao_3_tru(tru_A, tru_B, a, b, N_riemann)
```

```python
print("The tich phan giao nhau Riemann = ", f"{V_riemann_3tru:.6f}")
```
