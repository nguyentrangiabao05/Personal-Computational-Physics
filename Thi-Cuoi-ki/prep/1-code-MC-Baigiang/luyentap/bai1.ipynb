```python
N = 10000
a = 0
b = 1
```

# Phuong phap trung binh mau

```python
def ham_fx(x):
    f = np.sqrt(1-x**2)
    return f
```

```python
def MC_int(N, ham_fx, a, b):
    df = 0
    for i in range (N):
        x = np.random.uniform(a,b)
        df = df +  ham_fx(x)
    F = (b-a)/N * df
    return F
```

```python
F = MC_int(N, ham_fx=ham_fx, a=a, b=b)
```

```python
print(F)
```

```python
dem_trung = 0

for i in range(N):
    x = np.random.uniform(0, 1)
    y = np.random.uniform(0, 1)

    if x**2 + y**2 <= 1:
        dem_trung += 1

I = dem_trung / N

print("I gan dung =", I)
print("pi/4 =", np.pi / 4)
```

```python
# Tong Riemann
N_Riemann = N
x_arr = np.linspace(a, b, N_Riemann)
dx = x_arr[1] - x_arr[0]
dV = dx
```

```python
def tinhtong_Riemann(x_arr):
    F = 0
    for i in range(len(x_arr)):
        F = F + dV * ham_fx(x_arr[i])
    return F
```

```python
F_riemann  = tinhtong_Riemann(x_arr)
```
