```python
import numpy as np
import matplotlib.pyplot as plt
plt.style.use('sci.mplstyle')
import time
```

```python
def tinh_MC_quacau(N):
    global R
    tich_phan = 0
    for i in range(N):
        x  = np.random.uniform(-R, R)
        y  = np.random.uniform(-R, R)
        z =  np.random.uniform(-R, R)

        x_prime  = np.random.uniform(-R, R)
        y_prime  = np.random.uniform(-R, R)
        z_prime =  np.random.uniform(-R, R)

        if x**2 + y**2 + z**2 <= R**2 and x_prime**2 + y_prime**2 + z_prime**2 <= R**2: # chi lay nhung diem trong hinh tron
            dV = 1 / np.sqrt((x_prime - x)**2 + (y_prime - y)**2 + (z_prime - z)**2)
            tich_phan = tich_phan + dV
    
    V_hop_6D = (2*R)**6

    I = tich_phan * V_hop_6D / N

    return I
```

```python
R = 1
I = tinh_MC_quacau(10**6)
N = 10**5

V_cau = 4/3 * np.pi * R**3

I = tinh_MC_quacau(N) 

trungbinh_r_prime_minus_r = I/ V_cau**2
```
