```python
import numpy as np
import matplotlib.pyplot as plt
plt.style.use('sci.mplstyle')
import time
```

```python
# ham chi thi
def ellipsoid(x,y,z):
    global a, b, c
    f_ellipsoid = (x/a)**2 + (y/b)**2 + (z/c)**2
    if f_ellipsoid <= 1:
        return 1
    else:
        return 0
    
def tinh_MC_moment_khoiluong_chithi(N,a,b,c,rho,hinhdang):
    V_bao = 8*a*b*c
    dI_z = 0
    dM = 0
    for i in range(N):
        x = np.random.uniform(-a,a)
        y = np.random.uniform(-b,b)
        z = np.random.uniform(-c,c)

        dI_z = dI_z + rho * hinhdang(x,y,z) * (x**2+y**2)
        dM = dM + rho * hinhdang(x,y,z)

    I_z = dI_z * V_bao / N
    M = dM * V_bao / N
    return I_z, M    
```

```python
def tinh_MC_moment_khoiluong(N,a,b,c,rho,hinhdang):
    V_bao = 8*a*b*c
    dI_z = 0
    dM = 0
    for i in range(N):
        x = np.random.uniform(-a,a)
        y = np.random.uniform(-b,b)
        z = np.random.uniform(-c,c)

        if (x/a)**2 + (y/b)**2 + (z/c)**2 <= 1:
            dI_z = dI_z + rho * (x**2+y**2)
            dM = dM + rho

    I_z = dI_z * V_bao / N
    M = dM * V_bao / N
    return I_z, M

# 2 cach tuong duong
```

```python
N = 10**6
a = 2
b = 1
c = 1
rho = 1
```

```python
I_z, M = tinh_MC_moment_khoiluong(N,a,b,c,rho,ellipsoid)
```

```python
print("Ti so cua I/M (MC) = ", f"{I_z/M:.6f}")
print("Ti so cua I/M (Calculus) = ", f"{(1/5) * (a**2+b**2):.6f}")
```
