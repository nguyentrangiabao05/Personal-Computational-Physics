```python
N = 10**6
a = -1
b = 1
```

```python
def ham_fx(x,y):
    f = x**2+y**2
    return f
```

```python
def MC_int(N, ham_fx, a, b):
    N_in = 0
    for i in range(N):
        x = np.random.uniform(a,b)
        y = np.random.uniform(a,b)
        if ham_fx(x,y) <= 1:
            N_in = N_in +1
    return N_in
```

```python
N_in = MC_int(N, ham_fx, a, b)
pi = 4* N_in/N
```

```python
print("So pi = ", pi)
```

# Nhieu chieu

```python
import numpy as np

def tao_a_b(d):
    a = np.ones(d) * (-1)
    b = np.ones(d) * 1
    return a, b

def ham_fx_nhieuchieu(d, x):
    ham_fx_d_chieu = 0
    for i in range(d):
        ham_fx_d_chieu = ham_fx_d_chieu + x[i]**2

    return ham_fx_d_chieu

def tinhMC(d, N):
    a, b = tao_a_b(d)

    # x[i, j] = toa do thu i cua diem thu j
    x = np.ones((d, N))

    # The tich hinh hop [-1,1]^d
    V_hinhhop = 1
    for i in range(d):
        V_hinhhop = V_hinhhop * abs(b[i] - a[i])

    # Sinh ngau nhien N diem trong hinh hop d chieu
    for i in range(d):
        for j in range(N):
            x[i, j] = np.random.uniform(a[i], b[i])

    # Dem so diem nam trong hinh cau don vi
    N_in = 0
    for j in range(N):
        if ham_fx_nhieuchieu(d, x[:, j]) <= 1:
            N_in = N_in + 1

    # The tich hinh cau d chieu
    V_cau = V_hinhhop * N_in / N

    return V_cau, N_in

N = 1000000
d = 8
import math

V_cau, N_in = tinhMC(d, N)

giaitich  = (np.pi**(d/2))/((math.gamma(d/2 + 1)))

print("N_in =", N_in)
print("The tich uoc luong =", V_cau)
print("The tich giai tich  =", giaitich)
```
