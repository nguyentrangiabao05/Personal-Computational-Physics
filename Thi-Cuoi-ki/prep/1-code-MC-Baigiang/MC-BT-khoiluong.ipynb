````python
import numpy as np
import matplotlib.pyplot as plt
plt.style.use('sci.mplstyle')
import time
````

````python
def matdokhi(x,y,z):
    rho = x**2 + y**2 +z**2
    return rho

def tinhMC(N, a, b): #a, b la kich thich khoi hop
    d_mass = 0
    for i in range(N):
        x = np.random.uniform(a,b)
        y = np.random.uniform(a,b)
        z = np.random.uniform(a,b)
        d_mass = d_mass + matdokhi(x,y,z)
    mass = (np.abs(a-b))**3 * d_mass / N
    return mass
````

````python
N = 10000
a = 0
b = 1
````

````python
mass = tinhMC(N,a,b)
````

````python
print("Tinh MC - Khoi luong cua khoi khi la: ", mass)
````

# Tong Riemann


````python
N_Riemann = 100
dx = (b-a)/N_Riemann


x_arr = np.linspace(a + dx/2, b - dx/2, N_Riemann)
y_arr = np.copy(x_arr)
z_arr = np.copy(x_arr)

dV = dx**3
````

````python
def tong_Riemann(x_arr, y_arr, z_arr, dV):
    tong_khoiluong = 0
    for i in range(len(x_arr)):
        for j in range(len(y_arr)):
            for k in range(len(z_arr)):
                rho = matdokhi(x_arr[i],y_arr[j],z_arr[k])
                tong_khoiluong = tong_khoiluong + dV * rho
    return tong_khoiluong
````

````python
tong_Riemann = tong_Riemann(x_arr, y_arr, z_arr, dV)
````

````python
print("Tinh Riemann - Khoi luong cua khoi khi la: ", tong_Riemann)
````

# Giải thích vì sao code Riemann ban đầu bị lệch

Bài toán cần tính khối lượng:

$$
M=\int_0^1\int_0^1\int_0^1 (x^2+y^2+z^2)\,dxdydz
$$

Ta có thể tách tích phân:

$$
M=
\int_0^1 x^2dx+
\int_0^1 y^2dy+
\int_0^1 z^2dz
$$

Do đó:

$$
M=\frac{1}{3}+\frac{1}{3}+\frac{1}{3}=1
$$

Vậy giá trị đúng là:

$$
M=1
$$

## Vấn đề trong code Riemann ban đầu

Trong code ban đầu:

``` python
x_arr = np.linspace(a,b,N_Riemann)
dV = (x_arr[1]-x_arr[0])**3
```

Với `N_Riemann = 100`, ta có 100 điểm nhưng chỉ có 99 khoảng:

$$
dx=\frac{1}{99}
$$

Tuy nhiên vòng lặp lại cộng trên:

$$
100^3
$$

điểm lưới, mỗi điểm có thể tích:

$$
dV=\left(\frac{1}{99}\right)^3
$$

Tổng thể tích hiệu dụng:

$$
100^3\left(\frac{1}{99}\right)^3
=
\left(\frac{100}{99}\right)^3
\approx 1.03
$$

lớn hơn thể tích thật của khối lập phương:

$$
1^3=1
$$

nên kết quả bị lệch.

## Cách sửa đúng

Chia đoạn thành đúng 100 ô:

$$
dx=\frac{b-a}{N}
$$

và lấy điểm giữa mỗi ô:

``` python
dx = (b-a)/N_Riemann
x_arr = np.linspace(a + dx/2, b - dx/2, N_Riemann)
```

Khi đó:

$$
dV = dx^3
$$

và tổng thể tích:

$$
100^3\left(\frac{1}{100}\right)^3=1
$$

nên Riemann sum hội tụ đúng về giá trị thật.

-   Midpoint Riemann cho kết quả chính xác hơn.
-   Giá trị đúng của bài toán là:

$$
M=1
