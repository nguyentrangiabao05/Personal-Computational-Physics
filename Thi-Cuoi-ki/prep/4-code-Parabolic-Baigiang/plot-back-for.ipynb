```python
import numpy as np
import matplotlib.pyplot as plt

plt.style.use("sci.mplstyle")
```

```python
def ham_doc_file(filename):
    data = np.loadtxt(filename, comments="#")

    t_raw = data[:, 2]
    x_raw = data[:, 3]
    u_raw = data[:, 4]

    x = np.unique(x_raw)
    t = np.unique(t_raw)

    N_x = len(x)
    N_time = len(t)

    print("File:", filename)
    print("N_x =", N_x)
    print("N_time =", N_time)
    print("So diem =", len(u_raw))
    print("N_x * N_time =", N_x * N_time)

    if N_x * N_time != len(u_raw):
        raise ValueError("File khong phai luoi day du theo x va t")

    # file ghi theo:
    # for j in range(N_time):
    #     for i in range(N_x):
    temp = u_raw.reshape(N_time, N_x)

    return x, t, temp
```

```python
x_backward_1thanh , t_backward_1thanh, temp_backward_1thanh = ham_doc_file("truyen_nhiet_1thanh_backward_result.txt")
x_backward_2thanh , t_backward_2thanh, temp_backward_2thanh = ham_doc_file("truyen_nhiet_2thanh_backward_result.txt")
```

```python
plt.figure(figsize=(9, 6))

plt.imshow(
    temp_backward_1thanh,
    extent=[x_backward_1thanh.min(), x_backward_1thanh.max(), t_backward_1thanh.min(), t_backward_1thanh.max()],
    origin="lower",
    aspect="auto",
    cmap = "coolwarm"
)

plt.xlabel("Toa do (x)")
plt.ylabel("Thoi gian (t)")
plt.colorbar(label=" Nhiet do (x,t)")
plt.savefig("truyen_nhiet_1thanh_backward.png", dpi =300)
plt.show()
```

```python
plt.figure(figsize=(9, 6))

plt.imshow(
    temp_backward_2thanh,
    extent=[x_backward_2thanh.min(), x_backward_2thanh.max(), t_backward_2thanh.min(), t_backward_2thanh.max()],
    origin="lower",
    aspect="auto",
    cmap = "coolwarm"
)

plt.xlabel("Toa do (x)")
plt.ylabel("Thoi gian (t)")

plt.ylim(0,400)

plt.colorbar(label=" Nhiet do (x,t)")
plt.savefig("truyen_nhiet_2thanh_backward.png", dpi =300)
plt.show()
```

```python
fig = plt.figure(figsize=(9, 9))
ax = fig.add_subplot(111, projection='3d')

X_2thanh, T_2thanh = np.meshgrid(x_backward_2thanh, t_backward_2thanh)

ax.plot_surface(X_2thanh, T_2thanh, temp_backward_2thanh, color='white', edgecolor='gray', linewidth=0.3)
ax.set_xlabel('x')
ax.set_ylabel('t')
ax.set_zlabel('Temp')
ax.view_init(elev=20, azim=20)
plt.savefig("truyen_nhiet_2thanh_backward-3D.png", dpi=300)
plt.show()
```
