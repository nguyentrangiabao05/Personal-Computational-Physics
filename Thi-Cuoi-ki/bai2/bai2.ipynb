```python
import numpy as np
import matplotlib.pyplot as plt
plt.style.use('sci.mplstyle')
```

```python
def RK4(a, b, N, alpha, phuongtrinh):
    h = (b - a) / N                  # tinh kich thuoc buoc thoi gian
    x = np.linspace(a, b, N + 1)     # tao mang thoi gian tu a den b gom N+1 diem
    y = np.zeros((N + 1, len(alpha)))# tao mang nghiem cho he ODE
    y[0] = alpha                     # gan dieu kien dau

    for i in range(N):               # lap qua tung buoc thoi gian
        k1 = phuongtrinh(x[i], y[i]) # dao ham tai dau buoc
        k2 = phuongtrinh(x[i] + h/2, y[i] + h*k1/2) # dao ham tai diem giua dung k1
        k3 = phuongtrinh(x[i] + h/2, y[i] + h*k2/2) # dao ham tai diem giua dung k2
        k4 = phuongtrinh(x[i] + h, y[i] + h*k3)     # dao ham tai cuoi buoc

        y[i+1] = y[i] + h*(k1 + 2*k2 + 2*k3 + k4)/6 # cap nhat nghiem RK4

    return x, y                      # tra ve mang thoi gian va nghiem
```

```python
def hpt_bay_tu(t, Y):
    global q, m
    x  = Y[0] # Ham Y co cac gia tri x y z va van toc cua no
    y  = Y[1]
    z  = Y[2]
    vx = Y[3]
    vy = Y[4]
    vz = Y[5]

    Bx = -x*z # Tinh tu truong nhu cong thuc
    By = -y*z
    Bz = 1+z**2

    dvx_dt = q/m*(vy*Bz-vz*By) #tinh do thay doi van toc
    dvy_dt = q/m*(vz*Bx-vx*Bz)
    dvz_dt = q/m*(vx*By-vy*Bx)

    F = np.array([vx,vy,vz,dvx_dt,dvy_dt,dvz_dt]) #thi ham F la dY/dt -> dx dy dz va do thay doi van toc

    return F
```

```python
t_min = 0
t_max = 150

dt = 0.01

N_t = int((t_max-t_min)/dt + 1) # tinh so luong khoang chia N_t theo nhu dt, lay so nguyen

q = m = 1

x_0 = 0.1 #dieu kien ban dau
y_0 = 0.0
z_0 = 0.0

vx_0 = 0.08
vy_0 = 0.4
vz_0 = 0.15

Y0 = np.array([x_0, y_0, z_0, 
               vx_0,vy_0, vz_0]) # ghep cac dieu kien ban dau vo thanh 1 mang -> dua vo ham RK4
```

```python
def ghifile(t, Y, filename="baytu-RK4"): # luu file ket qua
    filename = f"KQ_{filename}.txt"

    with open(filename, "w", encoding="utf-8") as file:
        file.write("# Giai bai toan bay tu bang RK4\n")
        file.write("#\n")
        file.write(f"# {'t':>20s} {'x':>20s} {'y':>20s} {'z':>20s} {'v_x':>20s} {'v_y':>20s} {'v_z':>20s}\n")

        for t_step in range(len(t)):
            file.write(f"  {t[t_step]:20.8e} {Y[t_step, 0]:20.8e} {Y[t_step, 1]:20.8e} {Y[t_step, 2]:20.8e} {Y[t_step, 3]:20.8e} {Y[t_step, 4]:20.8e} {Y[t_step, 5]:20.8e}\n")
```

```python
t_baytu, Y_baytu = RK4(t_min, t_max, N_t, Y0, hpt_bay_tu) # call ham de giai
```

```python
ghifile(t_baytu, Y_baytu, filename="bai2-baytu-RK4") #luu vo file ket qua
```
