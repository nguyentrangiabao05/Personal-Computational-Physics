```python
import numpy as np
```

```python
def dieukienbien():
    global rho, A, N_z, N_t, z_max, L, t_max, T_mean, P, kappa, cp, Delta_T

    dz = z_max / (N_z - 1)
    dt = t_max / (N_t - 1)
    t = np.linspace(0, t_max, N_t)
    z = np.linspace(0, z_max, N_z)
    eta = kappa * dt / dz**2

    if eta > 0.5:
        print("Dieu kien on dinh khong duoc thoa man, hay tang N_t hoac giam N_z.")
        exit()

    T = np.zeros((N_z, N_t))

    T[:, 0] = T_mean + 0.03 * z
    T[0, :] = T_mean + Delta_T * np.sin(2 * np.pi * t / P)
    T[-1, :] = T_mean + 0.03 * z_max

    return T, z, t, dz, dt, eta
```

```python
def tinhtoan():
    T, z, t, dz, dt, eta = dieukienbien()

    for n in range(N_t - 1):
        for j in range(1, N_z - 1):
            T[j, n + 1] = T[j, n] + eta * (T[j + 1, n] - 2 * T[j, n] + T[j - 1, n]) + dt * A / (rho * cp)

        T[0, n + 1] = T_mean + Delta_T * np.sin(2 * np.pi * t[n + 1] / P)
        T[-1, n + 1] = T_mean + 0.03 * z_max

    return T, z, t, dz, dt, eta
```

```python
def luufile(T, z, t, output_name):
    global rho, A, N_z, N_t, z_max, L, t_max, T_mean, P, kappa, cp, Delta_T

    filename = f"Diavatly-ketqua-{output_name}.txt"

    with open(filename, "w", encoding="utf-8") as f:
        f.write("# KET QUA TINH TOAN DIA VAT LY\n\n")
        f.write("#" * 100 + "\n")
        f.write("# THAM SO DAU VAO\n")
        f.write("#" * 100 + "\n\n")
        f.write(f"# rho = {rho} kg/m^3\n")
        f.write(f"# A = {A} W/m^3\n")
        f.write(f"# N_z = {N_z}\n")
        f.write(f"# N_t = {N_t}\n")
        f.write(f"# z_max = {z_max} m\n")
        f.write(f"# L = {L} m\n")
        f.write(f"# t_max = {t_max} s\n")
        f.write(f"# T_mean = {T_mean} C\n")
        f.write(f"# P = {P} s\n")
        f.write(f"# kappa = {kappa} m^2/s\n")
        f.write(f"# cp = {cp} J/kg/K\n")
        f.write(f"# Delta_T = {Delta_T} C\n")
        f.write("\n")
        f.write("#" * 100 + "\n")
        f.write("# KET QUA\n")
        f.write("#" * 100 + "\n\n")
        f.write(f"#{'t':>15s} {'z':>15s} {'T':>15s}\n")

        for n in range(N_t):
            for j in range(N_z):
                f.write(f"{t[n]:15.5f} {z[j]:15.5f} {T[j, n]:15.5f}\n")
            f.write("\n")
```

```python
year = 365 * 24 * 3600

rho = 2800
A = 1e-6
N_z = 201
N_t = 4001
z_max = 35000
L = 35000
t_max = 1 * year
T_mean = 28
P = year
k = 2.5
cp = 800
Delta_T = 12
kappa = k / (rho * cp)
```

```python
T, z, t, dz, dt, eta = tinhtoan()
luufile(T, z, t, "FTCS")
```
